```python
# Transaction volumne each day!

data_df2_count = data_df2.groupby("date").count()
data_df2_count
```

```python
data_df2_count.describe()
```

```python
transaction_vol = data_df2_count['customer_id'].mean()
transaction_vol
```

```python
n_points = len(data_df2_count.index)
n_points
```

```python
# plotting :-

plt.figure(figsize=(20,10))
plt.plot(data_df2_count.index, data_df2_count['customer_id'],c='purple',
         label='Customer ID')

plt.plot(data_df2_count.index, np.linspace(transaction_vol,transaction_vol,
                                          n_points),c='r',
                                          label='Mean Transaction Volume')

plt.title("ANZ TRANSACTION VOLUME VS DATE",fontsize=20)
plt.xlabel("DATE", fontsize=20)
plt.ylabel("Number of customers",fontsize=20)
plt.legend(fontsize=20)
plt.tight_layout()
```

```python
months = [] 
for date in data_df['date']:
    if date.month == 8:
        months.append('August')
    elif date.month == 2 :
        months.append('February')
    elif date.month == 9:
        months.append("September")
    elif date.month == 10:
        months.append("October")

data_df['Months'] = months
data_df['Months'].head(50),data_df['Months'].tail(40)
```

```python
# Mean Customer Balance and payment amount by age:--

```

```python
df_cus_aug = data_df[data_df['Months'] == 'August'].groupby('customer_id').mean()

df_gen_aug = data_df[data_df['Months'] == 'August'].groupby('gender').mean()

```

```python
mean_f_bal_aug = df_gen_aug['balance'].iloc[0]
mean_m_bal_aug = df_gen_aug['balance'].iloc[1]

n_pts = len(df_cus_aug['age'])

mean_f_bal_aug, mean_m_bal_aug , n_pts
```

```python
plt.figure(figsize=(12,8))

plt.scatter(df_cus_aug['age'],
           df_cus_aug['balance'],
           c='purple',
           label='Balance')

plt.plot(df_cus_aug['age'],np.linspace(mean_f_bal_aug, mean_f_bal_aug,n_pts),
        c= 'y',
        label = 'Mean of Female Employee balances')

plt.plot(df_cus_aug['age'],np.linspace(mean_m_bal_aug, mean_m_bal_aug, n_pts),
        c = 'b',
        label = 'Mean of Male Employee balances')

plt.title('ANZ Customer Balances vs Age for August month',fontsize=20)

plt.xlabel('Age(years)', fontsize = 20)

plt.ylabel("Balance in $",fontsize = 20)

plt.legend()

plt.tight_layout()
```

```python
mean_f_amt_aug = df_gen_aug['amount'].iloc[0]
mean_m_amt_aug = df_gen_aug['amount'].iloc[1]

plt.figure(figsize=(12,8))

plt.scatter(df_cus_aug['age'],
           df_cus_aug['amount'],c = 'purple', label ='Amount')

plt.plot(df_cus_aug['age'],
            np.linspace(mean_f_amt_aug,mean_f_amt_aug,n_pts),
           c='yellow',
           label = "Mean Female Employee's amount")

plt.plot(df_cus_aug['age'],
        np.linspace(mean_m_amt_aug,mean_m_amt_aug,n_pts),
        c= 'r',
        label = "Mean Male Employee's amount")

plt.title("ANZ Customer Mean Payment Amount vs Age for August",fontsize=20)

plt.xlabel("Age in Years", fontsize = 20)

plt.ylabel("Amount in $",fontsize = 20)

plt.legend()

plt.tight_layout()
```
